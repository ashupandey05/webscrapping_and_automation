```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

option = webdriver.ChromeOptions()
option.add_argument("start-maximized")


driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=option)
driver.get('https://www.instagram.com/')
wait = WebDriverWait(driver, 5)
# lap1 = wait.until(EC.presence_of_element_located((By.NAME, "username")))
```

# Question 1

Login to your Instagram Handle

```python
#Login into the instagram account
username = driver.find_element(By.NAME, "username")
username.send_keys('SAMPLEUSERNAME')
password = driver.find_element(By.NAME, "password")
password.send_keys('SAMPLEPASSWORD')
password.submit()
```

# Question 2

Type for “food” in search bar and print all the names of the Instagram Handles that are displayed in list after typing “food”

```python
#Search bar is selected 
search_button = driver.find_elements(By.TAG_NAME, "svg")
search_button[2].click()
#type food in text area
text_area = wait.until(EC.presence_of_element_located((By.TAG_NAME, "input")))
text_area.send_keys('Food')
#extracting the list of displayed instagram handles
food_list = wait.until(EC.presence_of_all_elements_located((By.XPATH, '//div[starts-with(@class, "x9f619")]/a/div/div/div/div/div/div/div/span')))
for i in food_list:
    print(i.get_attribute('innerHTML'))
```

# Question 5

Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it

```python
search_button = driver.find_elements(By.TAG_NAME, "svg")
search_button[2].click()
text_area = wait.until(EC.presence_of_element_located((By.TAG_NAME, "input")))
text_area.send_keys('dilsefoodie')
select_handle = wait.until(EC.presence_of_element_located((By.XPATH, '//div[starts-with(@class, "x9f619")]/a/div/div/div/div/div/div/div/span')))
select_handle.click()
time.sleep(2)
#scrolling the page to get all the required post 
driver.execute_script('window.scrollBy(0, 1000);')
#storing the post as list
list_of_post = wait.until(EC.presence_of_all_elements_located((By.XPATH, '//article[@class = "x1iyjqo2"]/div/div/div/div')))  
#start the liking of 30 post
for i in range(30):
    list_of_post[i].click()
    lap12 = wait.until(EC.presence_of_all_elements_located((By.XPATH, '//section[contains(@class, "_ae48")]/span')))
    #getting the like button
    lap14=driver.find_elements(By.TAG_NAME,'svg')
    lap15=lap14[59].get_attribute('aria-label')
    if lap15=='Unlike':
        print("Already Liked!")
    else:
        lap12[0].click()
        print('post liked')
    driver.back()
```

# Question 8

Check the story of ‘coding.ninjas', print error messages accordingly -
If You have already seen the story.
Or The user has no story.
Or View the story if not yet seen

```python
search_button = driver.find_elements(By.TAG_NAME, "svg")
search_button[2].click()
text_area = wait.until(EC.presence_of_element_located((By.TAG_NAME, "input")))
text_area.send_keys('coding.ninjas')
select_handle = wait.until(EC.presence_of_element_located((By.XPATH, '//div[starts-with(@class, "x9f619")]/a/div/div/div/div/div/div/div/span')))
select_handle.click()
#selecting the story button 
story_button = wait.until(EC.presence_of_element_located((By.XPATH, '//section/main/div/header/div/div')))
#checking the if the user has posted story or not, if posted viewing the story 
if story_button.get_attribute('aria-disabled') == 'true':
    print('User has not posted any story')
else:
    print('User has posted story')
    view_story = driver.find_element(By.XPATH, '//div[@aria-disabled = "false"]')
    view_story.click()
```
